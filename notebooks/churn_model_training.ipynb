```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/Telco-Customer-Churn.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
drop_cols = [
    'CustomerID',
    'Churn Label',
    'Churn Score',
    'Churn Reason',
    'CLTV'
]

df.drop(columns=drop_cols, inplace=True)
```

```python
X = df.drop('Churn Value', axis=1)
y = df['Churn Value']
```

```python
y.value_counts()
```

```python
X.info()
```

```python
X.columns = X.columns.str.replace(' ', '_')
```

```python
X.isnull().sum().sort_values(ascending=False).head(10)
```

```python
X['Total_Charges'] = pd.to_numeric(X['Total_Charges'], errors='coerce')
X.dropna(inplace=True)
y = y.loc[X.index]
```

```python
X_encoded = pd.get_dummies(X, drop_first=True)
X_encoded.shape
```

```python
X_encoded.head()
```

```python
X_encoded.isnull().sum().sum()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred_lr = lr.predict(X_test)

print("Logistic Regression Report")
print(classification_report(y_test, y_pred_lr))
confusion_matrix(y_test, y_pred_lr)
```

```python
from sklearn.metrics import roc_auc_score

y_proba_lr = lr.predict_proba(X_test)[:, 1]
roc_auc_lr = roc_auc_score(y_test, y_proba_lr)

print("Logistic Regression ROC-AUC:", round(roc_auc_lr, 3))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight='balanced'
)

rf.fit(X_train, y_train)
```

```python
y_pred_rf = rf.predict(X_test)

print("Random Forest Report")
print(classification_report(y_test, y_pred_rf))
confusion_matrix(y_test, y_pred_rf)
```

```python
y_proba_rf = rf.predict_proba(X_test)[:, 1]
roc_auc_rf = roc_auc_score(y_test, y_proba_rf)

print("Random Forest ROC-AUC:", round(roc_auc_rf, 3))
```

```python
import pandas as pd

feature_importance = pd.Series(
    rf.feature_importances_,
    index=X_encoded.columns
).sort_values(ascending=False)

feature_importance.head(10)
```
